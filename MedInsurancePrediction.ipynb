DATA PREPROCESSING

```python
insurance_dataset.isnull().value_counts()
```

Encoding the categorical feature

```python
#Encoding Sex feature
insurance_dataset.replace({'sex':{'male':0,'female':1}},inplace=True)

#Encoding Smoker column
insurance_dataset.replace({'smoker':{'no':0,'yes':1}},inplace=True)

#Encoding region cloumn
insurance_dataset.replace({'region':{'southeast':0,'southwest':1,'northeast':2,'northwest':3}},inplace=True)
insurance_dataset
```

Splitting the features and target

```python
X = insurance_dataset.drop(columns='charges',axis=1) #dropping columns so axis will be 1
Y = insurance_dataset['charges']
```

```python
print(X)
```

```python
print(Y) #it contains charges coumns
```

Splitting the data into training data and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

Linear Regression

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train,Y_train)
```

Model Evaluation

```python
training_data_evaluation = regressor.predict(X_train)
```

```python
#r_squared value
r2_train = metrics.r2_score(Y_train, training_data_evaluation)
print("R squared value: ",r2_train)
```

```python
test_data_prediction = regressor.predict(X_test)
```

```python
r2_test = metrics.r2_score(Y_test ,test_data_prediction)
print("R squared value: ",r2_test)
```

```python
#Since the values in test and train are almost equal so they're not overfitting or overtraining each other
```

Building predictive system

```python
input_data = (31,1,25.74,0,0,0)

#Changing input tuple into numpy array
input_numpArr = np.asarray(input_data)

input_arrReshaped = input_numpArr.reshape(1,-1)
prediction = regressor.predict(input_arrReshaped)
print(prediction) #it will print in list or array format

print("The insurance cost in USD is: ",prediction[0])
```
